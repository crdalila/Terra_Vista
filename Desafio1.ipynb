```python
import pandas as pd
import random
import numpy as np

# Configuración
random.seed(42)
clientes = [f"Cliente_{i+1}" for i in range(40)]
tipos_cliente = ["Crecer", "Retención", "Neutro"]
tipos_tarea = ["Copy revision", "Design issue", "Requested change", "New item"]
prioridades = ["Urgente", "Alta", "Normal", "Baja"]

# Generador de texto de tareas con urgencia
def generar_texto(tipo_tarea, prioridad):
    base_textos = {
        "Copy revision": "Revisar y corregir el texto del cliente.",
        "Design issue": "Hay un problema con el diseño que debe resolverse.",
        "Requested change": "El cliente solicitó un cambio específico.",
        "New item": "Se necesita crear un nuevo recurso desde cero."
    }
    urgencias = {
        "Urgente": "Esto es muy urgente. Necesitamos resolverlo hoy.",
        "Alta": "Por favor, dale prioridad a esto.",
        "Normal": "Se puede hacer durante la semana.",
        "Baja": "No es prioritario, puede esperar unos días."
    }
    return f"{base_textos[tipo_tarea]} {urgencias[prioridad]}"

# Generador de datos
data = []
for i in range(2000):  # ✅ Generar 2000 registros
    cliente = random.choice(clientes)
    tipo_cliente = random.choices(tipos_cliente, weights=[0.3, 0.4, 0.3])[0]
    tipo_tarea = random.choice(tipos_tarea)

    # Regla para prioridad: urgente más probable en tareas críticas y clientes "crecer"
    if tipo_cliente == "Crecer" and tipo_tarea in ["Design issue", "Requested change"]:
        prioridad = random.choices(prioridades, weights=[0.4, 0.3, 0.2, 0.1])[0]
    elif tipo_cliente == "Retención":
        prioridad = random.choices(prioridades, weights=[0.2, 0.4, 0.3, 0.1])[0]
    else:
        prioridad = random.choices(prioridades, weights=[0.1, 0.2, 0.4, 0.3])[0]

    texto = generar_texto(tipo_tarea, prioridad)

    data.append({
        "cliente": cliente,
        "tipo_cliente": tipo_cliente,
        "tipo_tarea": tipo_tarea,
        "prioridad": prioridad,
        "texto_tarea": texto
    })

# Crear DataFrame
df = pd.DataFrame(data)

# Guardar como CSV
df.to_csv("datos_tareas_ficticias_2000.csv", index=False)
print("CSV generado con 2000 registros: datos_tareas_ficticias_2000.csv")
```
