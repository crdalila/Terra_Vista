### Problema a resolver

Al presentarse un problema poco definido y nada claro, buscamos diferentes necesidades que podría tener el cliente. En este caso, nos centramso en resolver dos.

- El tipo de problema en dónde se clasificará el request del cliente sobre la página web 

+ La prioridad de dicho problema como alta, media  baja en base a diferentes factores. 

Para ello, en este primer data set podemos rescatar del cuerpo del texto cierta información que nos detalle qué tipo de prioridad se presenta y que prioridad tendrá en el proceso de resolverla.

#### GENERACIÓN DE DATOS SINTÉTICOS 

Al existir falta de datos y tener un limitado acceso a datos similares en el mercado, hemos optado por crear datos sintéticos en basea tres criterios importantes:
 
 + Criterios de prioridad: La importancia que tiene el problema en sí basado en diferentes parámetros que afectarán a esa importancia. Estos prámetros se basarán en importancia del cliente,fidelidad, proyección externa, dimensión del cliente, etc. 
 
 Los cuatro términos finales en los que se calificarán cada problema o request serán: 

        Low
        Medium
        High
        Critical 

- Duración: En base al problema o request presentad y con un entrenamiento e investigación de un histórico de requests similares, podemos crear un modelo en el que se estime la duración que tendrá ese trabajo concreto para el equipo de Terra y con ello, planear y organizar calendarios laborales mucho más claros y concisos.

* Tipo de tarea: Mismo camino que la duración. En base a históricos y similitud ente problemas, podemos agruparlos en tipos concretos para poder redirigir mucho mejor el flujo de tabajo hacia los departamentos corresponientes además de facilitar mucho el ritmo de tabajo y la organización de los mismos datos.

#### DESARROLLO DE VARIABLES Y ORGANIZACIÓN DE DATASETS

Se plantean dos datasets. Uno para lograr una categoría numéica del 1 al 10 en base a la prioridad del cliente. esto es:
Dependiendo de variables como:

 **Tamaño de empresa**, **Tamaño de proyecto**, **Facturación histórica** , **interés** e **Tasa de conversión** (que será un porcentaje). Cada uno de ellos, tendrá un peso establecido marcado por Terra basado en la importancia que se quiera dar a cada variable. Con una ponderación, desarrolamos los pesos de cada variable para definir su importancia. 



Con esto, nosotros sacamos un número del 0 al 1 reflejando la PRIORIDAD BASE DE LA EMPRESA/CLIENTE con la que vamos a empezar a trabajar. 



#### propuesta de pesos, ejemplo,
| **Variable**           | **Peso sugerido (%)** | **Justificación breve**                          |
|------------------------|-----------------------|--------------------------------------------------|
| Facturación            | 25%                   | Muestra poder económico.                         |
| Tamaño de empresa      | 15%                   | Afecta capacidad operativa, pero menos crítico.  |
| Tamaño del proyecto    | 25%                   | Valor directo del negocio potencial.             |
| Potencial de empresa   | 20%                   | Crecimiento futuro, muy importante.              |
| Tasa de conversión     | 15%                   | Refleja probabilidad de éxito real.              |


**La fórmula de Prioridad** combina cinco variables clave del negocio, cada una ponderada según su importancia estratégica. El objetivo es calcular un índice de prioridad que permita ordenar y seleccionar oportunidades o clientes de forma objetiva.

        Prioridad=0,25⋅F+0,15⋅E+0,25⋅P+0,20⋅C+0,15⋅T
        

F = Facturación (normalizada)
E = Tamaño de empresa (normalizado)
P = Tamaño del proyecto (normalizado)
C = Potencial de la empresa (normalizado)
T = Tasa de conversión (valor entre 0 y 1 o porcentaje normalizado)
