#### **DESARROLLO DE VARIABLES Y ORGANIZACIÓN DE DATASETS**


**Se plantean dos datasets.**
Hablemos del primero: Creado sintéticamente basado en empresas que pueden ser potenciales clientes de Terra para lograr una categoría numérica del 1 al 5 en base a la prioridad del cliente. esto es:

Dependiendo de variables como:

 **Industry, GrowthPotential, CulturalFit, Reputation, Profitability y LongTermRelationship** y en base a estos datos, se generará una columna nueva llamada **ClientStrategicFit** (que será un porcentaje). Cada uno de ellos, tendrá un peso establecido marcado por Terra basado en la importancia que se quiera dar a cada variable. Con una ponderación, desarrolamos los pesos de cada variable para definir su importancia. 



Con esto, nosotros sacamos un número del 1 al 5 reflejando la PRIORIDAD BASE DE LA EMPRESA/CLIENTE con la que vamos a empezar a trabajar. 




| **Variable**           | **Peso sugerido (%)** | **Justificación breve**                          |
|------------------------|-----------------------|--------------------------------------------------|
| Industry            | 10%                   | Sector del cliente                         |
| GrowthPotential     | 20%                   | Crecimiento futuro, muy importante  |
| CulturalFit    | 15%                   | Compatibilidad entre las culturas del cliente y de la agencia             |
| Reputation   | 15%                   |     Reputación del cliente     |
| Profitability     | 15%                   | Rentabilidad de la empresa              |
|LongTermRelationship    |    15%      |   El valor esperado en una relación a largo plazo entre empresa y Terra  |
| ClientPortfolioWeight |    10% |  El peso relativo de cada cliente sobre el total de la cartera |                   


**La fórmula de Prioridad** combina seis variables clave del negocio, cada una ponderada según su importancia estratégica. El objetivo es calcular un índice de prioridad que permita ordenar y seleccionar oportunidades o clientes de forma objetiva.
El peso de cada variable sobre el índice depende de la etapa en la que se encuentre la empresa y del sector, y pueden ser ajustados caso a caso.


$$
\text{Prioridad} = 0.1 \cdot I + 0.2 + 0.15 \cdot GP + 0.15 \cdot R + 0.15 \cdot P + 0.15 \cdot LTR + 0.1 \cdot CPW
$$


        

Se aplica el ClientStrategicFit tras predecir la prioridad desde el request_text que introduce el cliente. De esta forma se ajusta el valor predicho al valor de prioridad ponderada respecto al interés que se tiene del cliente. Así, una tarea de alta prioridad puede pasar a urgente o normal dependiendo de la importancia del cliente y su proyecto.

#### Modelo elegido y desarrollo del Machine Learning



Creando un Pipeline que encadena dos pasos principales:

**TfidfVectorizer(stop_words='english')**

Convirtiendo el texto en números que el modelo puede entender, usando la técnica TF-IDF (Term Frequency-Inverse Document Frequency). Esta técnica da más importancia a las palabras que aparecen mucho en un documento pero poco en los demás.

Aquí se eliminan las palabras comunes del inglés (stop_words='english').

**LogisticRegression(max_iter=1000)**

Comenzamos Usando un clasificador estadístico que aprende a separar las clases basándose en los valores de entrada (en este caso, los TF-IDF), asegurándonos que el modelo tenga suficiente tiempo para converger si los datos son complejos.

#### Modelo seleccionado_ XGBoost

Empezamos con un pipeline que combina tres componentes clave:

1. **LabelEncoder (Codificación de etiquetas)**
Convertimos las etiquetas categóricas en números que el modelo necesita para funcionar, y se usa también para decodificar las predicciones después.

2. **TfidfVectorizer(stop_words='english')**
Transformamos el texto en vectores numéricos usando la técnica TF-IDF, también con *stop_words='english'* para evitar ruido y para conviertir el texto en un formato que el modelo XGBoost puede procesar.

3.**XGBClassifier()**
Es un modelo de clasificación basado en XGBoost (eXtreme Gradient Boosting), un algoritmo muy potente que combina muchos árboles de decisión. 

**PROCESO COMPLETO:**
**Codificación de etiquetas:** LabelEncoder transforma y_train y y_test en valores numéricos.

**Construcción del pipeline:** TfidfVectorizer para representar los textos y XGBClassifier como clasificador.

**Entrenamiento:** el pipeline se entrena con X_train y y_train_encoded.

**Predicción:** se generan prediccionesy se descodifican con LabelEncoder para interpretarlas como etiquetas originales.

**Evaluación:** se genera un classification_report que incluye:

+ Precisió; qué porcentaje de predicciones fueron correctas.

- Exhaustividad; cuántas verdaderas se encontraron.

- F1-score; equilibrio entre precisión y exhaustividad para cada clase.

#### METODOLOGÍA

Se utilizó un pipeline de Scikit-learn compuesto por:

```python
Pipeline([
    ("tfidf", TfidfVectorizer(max_features=5000, stop_words="english")),
    ("regressor", RandomForestRegressor(n_estimators=100, max_depth=10, random_state=42))
])
```



    TfidfVectorizer transforma el texto de la solicitud en una representación numérica basada en la frecuencia de los términos.

    RandomForestRegressor se eligió por su capacidad para capturar relaciones no lineales y su robustez frente a ruido.

Cada pipeline fue entrenado y evaluado por separado, adaptándose a las características específicas de cada tipo de solicitud.

### **FEATURE ENGINEERING**

Lo primero de todo es hacer limpieza y normalización de los datos. Pero en un primer momento decidimos utilizar solamente el request, el cuerpo del texto donde el cliente explica el problema que el encuentra en el entregable. 


a. Eliminación de columnas irrelevantes
Se eliminaron las columnas que no aportaban valor directo al objetivo de clasificación, como:

Page, Input Date, Status, Browser, Device, Estimated Time (tokens)

Posteriormente, también se eliminó Company para centrarse exclusivamente en el texto de la solicitud (Request).

 #### **Selección de variables**

Variable dependiente (y): se seleccionó la columna Priority, que representa la clase a predecir (Urgent, High, Normal, Low).



#### Construcción del pipeline de NLP + ML


Se construyó un Pipeline de Scikit-learn compuesto por:

**TfidfVectorizer():** para transformar el texto en una matriz numérica basada en la frecuencia de términos ponderada por su rareza.

**RandomForestClassifier()**: con parámetros personalizados para limitar la complejidad y evitar sobreajuste.

**Justificación:** Se optó por un pipeline para encapsular el preprocesamiento y el modelo, facilitando la validación cruzada futura y posibles mejoras. Se eligió Random Forest por su robustez frente a ruido y facilidad de interpretación.
