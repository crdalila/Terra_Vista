#### GENERACIÓN DE DATOS SINTÉTICOS 

Al existir falta de datos y tener un limitado acceso a datos similares en el mercado, hemos optado por crear datos sintéticos en basea tres criterios importantes:
 
 + Criterios de prioridad: La importancia que tiene el problema en sí basado en diferentes parámetros que afectarán a esa importancia. Estos prámetros se basarán en importancia del cliente, fidelidad, proyección externa, dimensión del cliente, etc. 
 
 Los cuatro términos finales en los que se calificarán cada problema o request serán: 

        Low
        Medium
        High
        Critical 

- Duración: En base al problema o request presentad y con un entrenamiento e investigación de un histórico de requests similares, podemos crear un modelo en el que se estime la duración que tendrá ese trabajo concreto para el equipo de Terra y con ello, planear y organizar calendarios laborales mucho más claros y concisos.

* Tipo de tarea: Mismo camino que la duración. En base a históricos y similitud ente problemas, podemos agruparlos en tipos concretos para poder redirigir mucho mejor el flujo de tabajo hacia los departamentos corresponientes además de facilitar mucho el ritmo de tabajo y la organización de los mismos datos.

#### DESARROLLO DE VARIABLES Y ORGANIZACIÓN DE DATASETS


**Se plantean dos datasets.**
Hablemos del primero: Creado sintéticamente basado en empresas que pueden ser potenciales clientes de Terra para lograr una categoría numérica del 1 al 5 en base a la prioridad del cliente. esto es:

Dependiendo de variables como:

 **Industry, GrowthPotential, CulturalFit, Reputation, Profitability y LongTermRelationship** y en base a estos datos, se generará una columna nueva llamada **ClientStrategicFit** (que será un porcentaje). Cada uno de ellos, tendrá un peso establecido marcado por Terra basado en la importancia que se quiera dar a cada variable. Con una ponderación, desarrolamos los pesos de cada variable para definir su importancia. 



Con esto, nosotros sacamos un número del 1 al 5 reflejando la PRIORIDAD BASE DE LA EMPRESA/CLIENTE con la que vamos a empezar a trabajar. 



#### propuesta de pesos, ejemplo,
| **Variable**           | **Peso sugerido (%)** | **Justificación breve**                          |
|------------------------|-----------------------|--------------------------------------------------|
| Industry            | 20%                   | Sector del cliente                         |
| GrowthPotential     | 20%                   | Crecimiento futuro, muy importante  |
| CulturalFit    | 10%                   | Compatibilidad entre las culturas del cliente y de la agencia             |
| Reputation   | 20%                   |     Reputación del cliente     |
| Profitability     | 15%                   | Rentabilidad de la empresa              |
|LongTermRelationship    |    15%      |   El valor esperado en una relación a largo plazo entre empresa y Terra  |


**La fórmula de Prioridad** combina seis variables clave del negocio, cada una ponderada según su importancia estratégica. El objetivo es calcular un índice de prioridad que permita ordenar y seleccionar oportunidades o clientes de forma objetiva.

        Prioridad=0,25⋅F+0,15⋅E+0,25⋅P+0,20⋅C+0,15⋅T

### DATASET FINAL

Con datos sintéticos en los que se vgeneran 10000 filas con estas features: 

* Company: El nmbre de la empresa

* Input Date: Fecha del input (request)

+ Device: Dispositivo electrónico que se usa para enviar esa petición

+ Browser: Buscador utilizado

+ Page: Enlace (URL) a la página web en construcción
 
+ Estimated Time (tokens): tiempo estimado que se considera que esa tarea necesita para ser concluida

+ Real Time: Tiempo real que esa tarea ha necesitado para ser concluida

+ ClientStrategicFit: La métrica sintética que aparece del primer dataset realizado para representar la medida de compatibilidad entre el cliente/empresa y Terra.
