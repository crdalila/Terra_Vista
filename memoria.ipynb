## MEMORIA PROYECTO TERRA VISTA 

### Análisis y estructura del proyecto.

Vamos a coordinar y estructurar el trabajo y el órden claro de los pasos a seguir. 

Empezamos con un estudio exhaustivo de la necesidad primordial de Terra y, en base a ella, desarrollar una estrategia de trabajo buscando dónde encajar nuestra vertical y las habilidades de data. 

### BÚSQUEDA DE DATOS REALES Y ÚTILES 

A través de esta [página_web]('https://www.kaggle.com/datasets/tobiasbueck/capterra-reviews'), sacamos un primer data set de datos reales que tienen tres columnas que nos interesan: 

+ BODY: El problema que comunica el cliente con, en este caso, el sistema IT. 

+ TYPE:  El tipo de problema que refleja el texto (body)

- PRIORITY: La prioridad de dicho problema. 


Con este dataset hacemos una primera aproximación a crear un modelo de predicción en base al NLP que vectorizando el texto, sea capaz de predecir qué tipo de categoría es cada problema. 

#### API DATA 

A través de flask, hemos hecho dos end-points para categoría y para prioridad que, de momento, devuelven una categoría  o prioridad aleatoria.


Esta API la consumirá fs team para su proyecto web.
