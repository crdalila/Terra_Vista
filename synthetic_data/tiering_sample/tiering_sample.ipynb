```python
import pandas as pd
```

```python


# Datos previos
data = [
    ["Airbnb", 5, 4, 4, 5, 5, 5],
    ["UN", 2, 2, 3, 3, 1, 5],
    ["Abbott", 4, 4, 4, 4, 4, 5],
    ["Healthline Media", 4, 4, 4, 4, 3, 4],
    ["Talkdesk", 4, 5, 4, 4, 2, 4],
    ["Stifel Bank", 3, 4, 3, 3, 4, 4],
    ["DAT", 3, 3, 3, 3, 3, 4],
    ["Ruby's Rainbow", 1, 1, 4, 2, 1, 3],
    ["SEI", 4, 5, 4, 4, 4, 4],
    ["New York Hall of Science", 2, 2, 3, 3, 1, 4],
    ["USC", 3, 3, 4, 4, 3, 5],
    ["Essilor Luxottica", 5, 4, 4, 5, 5, 5],
    ["Healthgrades", 3, 3, 3, 3, 3, 3],
    ["Lucet", 3, 3, 3, 2, 2, 3],
    ["Valorx", 4, 3, 4, 4, 3, 4],
    ["Unqork", 4, 5, 4, 4, 2, 4],
    ["nTopology", 4, 5, 4, 4, 2, 4],
    ["Rev Clinics", 2, 2, 3, 2, 3, 3],
    ["Dental Sleep Solutions", 2, 2, 3, 2, 3, 3],
    ["Cardata", 4, 4, 3, 3, 3, 4],
    ["Transcent", 3, 3, 3, 3, 3, 3],
    ["Telink", 5, 5, 4, 4, 4, 5],
    ["AXA", 5, 4, 4, 5, 5, 5]
]

columns = [
    "Company", "Industry", "GrowthPotential", "CulturalFit",
    "Reputation", "Profitability", "LongTermRelationship"
]

df = pd.DataFrame(data, columns=columns)

# Calcular Client Strategic Fit con ponderaciones
#media ponderada para Terra con el objetivo de consolidar  
weights = {
    "Industry": 0.20,
    "GrowthPotential": 0.20,
    "CulturalFit": 0.10,
    "Reputation": 0.20,
    "Profitability": 0.15,
    "LongTermRelationship": 0.15
}

df["ClientStrategicFit"] = (
    df["Industry"] * weights["Industry"] +
    df["GrowthPotential"] * weights["GrowthPotential"] +
    df["CulturalFit"] * weights["CulturalFit"] +
    df["Reputation"] * weights["Reputation"] +
    df["Profitability"] * weights["Profitability"] +
    df["LongTermRelationship"] * weights["LongTermRelationship"]
).round(2)

# Guardar como CSV
csv_path = "client_strategic_fit.csv"
df.to_csv(csv_path, index=False)

csv_path
```
