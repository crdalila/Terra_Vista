### DIAPO 3

**Modelo de duración**

Primer partimos del modelo predictor del tipo de tarea, para después crear cuatro modelos por cada tarea para poder determinar el tiempo exacto de duraci´n que conllevará realizar dicha tarea. 

Nos hemos dado cuenta de que cada tarea tiene una duración estimada diferente a las otras, por tanto es importante entrenar cada modelo con tiempos diferentes, más o menos altos para ajustar lo mejor posible el tiempo de duración de cada tarea.

Se clasifica por horas de duración que se tardará en terminar cada tarea. 


**Objetivo**

- Descargar trabajo al mánager

+ Organizar las tareas en base a su duración de resolución

### DIAPO 5

**Mantenimiento del modelo y carga**

Cómo quedaría el sistema de mantenimiento y reentrenamiento (actualización) del modelo.

De la aplicación Terra Vista, se pasa unos campos (unos inputs) a Click Up, algunos generads automáticamente, otros por el usuario...

De la misma aplicación recibimos el request del cliente En formato texto que pasará a nuestra API predictora y que devolverá campos para consumir en Click Up. 

Desde click up recolectamos datos para alimentar a nuestro modelo, por un lado tendremos un hisórico dedatos que iremos almacenando todos los registros (como una wardrobe), por otro, tendremos la información del cliente que usaremos para ajustar las prioridades de las tareas. Con esta infrmación reentreno y mantenemos el modelo actualizado y al día.
